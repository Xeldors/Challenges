```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###Pi with Monte Carlo

radius=10
sie=500000

##Define the Random Uniform constrained between -1 and 1

uniform_values_1 = np.random.uniform(low=-radius, high=radius, size=sie)
uniform_values_2 = np.random.uniform(low=-radius, high=radius, size=sie)

#Dataset
UniformValues= pd.DataFrame({"x":uniform_values_1,"y":uniform_values_2})

#Define if the point is inside the circle
Distance=[]
for i in range(0, len(UniformValues)-1):
    dist=np.sqrt(float(UniformValues["x"].loc[i])**2 + float(UniformValues["y"].loc[i])**2)
    Distance.append(dist)

UniformValues=pd.concat([UniformValues,pd.DataFrame({"Distance":Distance})],axis=1)

AreaPoints=[]
for i in range(0, len(UniformValues)-1):
    if Distance[i]<radius: 
        AreaPoints.append(1)
    else: 
        AreaPoints.append(0)
        

Proportion= sum(AreaPoints)/sie
squareearea=((radius - (-radius))**2)
AreaCircle=Proportion*squareearea
                       
UniformValues=pd.concat([UniformValues,pd.DataFrame({"AreaPoints":AreaPoints})],axis=1)

Pi= AreaCircle/radius**2




plt.figure(figsize=(8, 6))
# Plot points that should not be highlighted

plt.scatter(UniformValues["x"][UniformValues["AreaPoints"] == 0], UniformValues["y"][UniformValues["AreaPoints"] == 0],
            c='blue', marker='o', edgecolor='k', label='Normal Points')

# Plot points that should be highlighted
plt.scatter(UniformValues["x"][UniformValues["AreaPoints"] == 1], UniformValues["y"][UniformValues["AreaPoints"] == 1],
            c='red', marker='o', edgecolor='k', label='Normal Points')
# Add titles and labels
plt.title('Scatter Plot of Uniform Values')
plt.xlabel('Uniform Values 1 (X)')
plt.ylabel('Uniform Values 2 (Y)')

Pi
```
